```python
import numpy
#Generate random actual and predicted values
actual  = numpy.random.binomial(n=1, p=0.9, size=1000)
predicted = numpy.random.binomial(n=1, p=0.9, size=1000)
```

```python
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(actual, predicted)
print(confusion_matrix)
```

```python
import matplotlib.pyplot as plt

#Display confusion matrix as a plot
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])


cm_display.plot()
cm_display.ax_.set_title("Confusion Matrix")
# Add text labels to each cell
cm_display.ax_.text(0, 0, '\nTrue Negative\n(TN)', ha='center', va='top', fontsize=8, color='white')
cm_display.ax_.text(1, 0, '\nFalse Positive\n(FP)', ha='center', va='top', fontsize=8, color='white')
cm_display.ax_.text(0, 1, '\nFalse Negative\n(FN)', ha='center', va='top', fontsize=8, color='white')
cm_display.ax_.text(1, 1, 'True Positive\n(TP)', ha='center', va='top', fontsize=8, color='black')

plt.show()
```

```python
#Accuracy measures how often the model is correct.

accuracy = metrics.accuracy_score(actual, predicted)

#Of the positives predicted, what percentage is truly positive?
precision = metrics.precision_score(actual, predicted)

#Sensitivity measures the proportion of actual positives that are correctly identified as such.
sensitivity = metrics.recall_score(actual, predicted)

#Specificity measures the proportion of actual negatives that are correctly identified as such.
specificity = metrics.recall_score(actual, predicted, pos_label=0)

#F1 Score is the harmonic mean of precision and recall.
f1_score = metrics.f1_score(actual, predicted)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Sensitivity:", sensitivity)  
print("F1 Score:", f1_score)
```

```python
def categorize_score(score):
    if score < 0.5:
        return "Bad", "\033[91m"  # Red
    elif score < 0.7:
        return "Poor", "\033[93m"  # Yellow
    elif score < 0.8:
        return "Good", "\033[92m"  # Green
    elif score < 0.9:
        return "Better", "\033[94m"  # Blue
    else:
        return "Best", "\033[95m"  # Magenta

reset = "\033[0m"  # Reset color

category, color = categorize_score(accuracy)
print(f"Accuracy: {color}{accuracy:.4f} ({category}){reset}")

category, color = categorize_score(precision)
print(f"Precision: {color}{precision:.4f} ({category}){reset}")

category, color = categorize_score(sensitivity)
print(f"Sensitivity: {color}{sensitivity:.4f} ({category}){reset}")

category, color = categorize_score(f1_score)
print(f"F1 Score: {color}{f1_score:.4f} ({category}){reset}")
```
