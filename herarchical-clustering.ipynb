```python
import numpy as np
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19,24, 17,16, 25,24, 22, 21, 21]

plt.scatter(x, y, color='blue', marker='o')
plt.title('Scatter Plot of Given Data Points')
plt.show()
```

```python
#Compute the ward linkage using euclidean distance and visualize the dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage

data = list(zip(x, y))

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.title('Dendrogram using Ward Linkage')
plt.show()
```

```python
#Same using sklearn
from sklearn.cluster import AgglomerativeClustering

hierarchical_cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = hierarchical_cluster.fit_predict(data)
print("Cluster labels for each point:", labels)

plt.scatter(x, y, c=labels)
plt.title('Agglomerative Clustering Result')
plt.show()
```
