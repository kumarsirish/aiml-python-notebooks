```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3,1,100)
y = numpy.random.normal(150, 40, 100) /x  #Creates inverse relation with x

plt.scatter(x,y)
plt.title("Complete Data")
plt.xlabel("Time spent on mall")
plt.ylabel("Amount spent on mall")
plt.show()

```

```python
#Split into training and testing data
train_x = x[:80]
train_y = y[:80]    
test_x = x[80:]
test_y = y[80:]

plt.scatter(train_x, train_y)
plt.title("Training Data")
plt.xlabel("Time spent on mall")
plt.ylabel("Amount spent on mall")
plt.show()  
```

```python
plt.scatter(test_x, test_y)
plt.title("Testing Data")
plt.xlabel("Time spent on mall")
plt.ylabel("Amount spent on mall")
plt.show()
```

```python
#Draw polynomial regression line

poly_model = numpy.poly1d(numpy.polyfit(train_x, train_y, 4)) #4 degree polynomial x^4+x^3+...
myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot( myline,poly_model(myline), color='red')
plt.title("Polynomial Regression Fit on Training Data")
plt.text(4, 300, 'Degree 4 Polynomial Fit', color='red')
plt.xlabel("Time spent on mall")
plt.ylabel("Amount spent on mall")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(train_y, poly_model(train_x))
print("R2 score for training data:", r2) #Higher the R2 score, better the model fits the data
```

```python
#Test with the testing data
r2 = r2_score(test_y, poly_model(test_x))
print("R2 score for testing data:", r2) #Higher the R2 score, better the model fits the data
```

```python
#Predict if a person spends 5 hours in mall, how much they will spend
hours = 5
amount_spent = poly_model(hours)
print("If a person spends", hours, "hours in mall, they will spend approximately $", round(amount_spent,2))
```
