```python
import matplotlib.pyplot as plt

x = [4,5,10,4,3,11,14,6,10,12]
y=[21,19,24,17,16,25,24,22,21,21]

plt.scatter(x, y)
plt.show()
```

```python
#Utilize the elbow method to visualize the intertial for different values of K
from sklearn.cluster import KMeans

data = list(zip(x, y))
intertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    intertias.append(kmeans.inertia_)

plt.plot(range(1,11), intertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Intertia')
plt.show()
```

```python
#We can see that the "elbow" on the graph above (where the interia becomes more linear) is at K=2. We can then fit our K-means algorithm one more time and plot the different clusters assigned to the data:

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show
```
