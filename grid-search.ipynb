```python
from sklearn import datasets
iris = datasets.load_iris()

print(iris.data[:5])  # Print the first 5 samples
print(iris.target[:5])  # Print the first 5 target values
X = iris['data']  #Metric, multi dimensional
y = iris['target'] #Vector one dimensional
```

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression(max_iter=1000)
print(logit.fit(X, y))
print(logit.score(X, y))
```

```python
# setting the values of hyper-parameter C takes domain knowledge and practice
C = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0]

scores = []

for choice in C:
    logit.set_params(C=choice)
    logit.fit(X, y)
    scores.append(logit.score(X, y))
print(scores)
```

We see that Model performed worst at C=1 , score incrased for higher value of C 1.75 but after that score it not approve.
